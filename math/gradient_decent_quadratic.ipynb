```python
# 梯度下降算法
# 求函数f(x)的最小值，最小值的点(0,2)
import numpy as np
#原函数
def f(x):
    return x**2 + 2
#f(x)的导数
def d(x):
    return 2*x

X = []     #用来存储x的变化趋势值
Y = []     #用来存储y的变化趋势值
```

```python
x = 2
lrate = 0.8    #学习率，步长
fv = 1         #初始化一个浮动值
y = f(x)        
X.append(x)     #把x和y值都假如到列表，后面绘图用
Y.append(y)
while fv > 1e-10:     #设置停止条件,  当梯度下降迭代到y几何不改变的时候停止迭代
    x = x - lrate * d(x)       #更新参数
    fv = np.abs(y - f(x))      #计算最新的y的变化值
    y = f(x)                   
    X.append(x)          #x的逐步变化情况
    Y.append(y)          #y的逐步变化情况
```

```python
print(x,y)   #x趋向于0， y趋向于2
print(X)
print(Y)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
X2 = np.arange(-2.,2.,0.05)
Y2 = X2**2 + 2
plt.plot(X2,Y2,'-',linewidth=2)
plt.plot(X,Y, 'bo--')
plt.grid(True)
plt.title(u'$y=x^2+2$ min value is x=%.2f,y=%.2f'%(x,y))
plt.show()
```
