```python
# 用python做二元二次方程的梯度下降实现
#原函数
# f的最小值是（0，0，0）
f = lambda x,y: x**2 + y**2 
#导数，x的导数2x，y的导数2y
d = lambda t: 2*t

X, Y, Z = [], [], []
x, y = 2, 2   #x和y初始值从2开始
fv = 1
z = f(x,y)    
lrate = 0.1    #步长，学习率
X.append(x)    #记录x的变化情况，绘图用
Y.append(y)
Z.append(z)
```

```python
while fv >1e-10:       #设定退出条件，当z值的变化很小时，停止循环
    x -= lrate*d(x)    #更新x值
    y -= lrate*d(y)    #更新y值
    fv = z - f(x,y)
    z = f(x,y)
    X.append(x)    #记录x的变化情况，绘图用
    Y.append(y)
    Z.append(z)
```

```python
print(x,y,z)
print(X,Y,Z)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig = plt.figure()
axes = Axes3D(fig)
X2 = np.arange(-2,2,0.2)
Y2 = np.arange(-2,2,0.2)
X2,Y2 = np.meshgrid(X2,Y2)
Z2 = X2**2 + Y2**2 
axes.plot_surface(X2,Y2,Z2,rstride=1,cstride=1,cmap='rainbow',zorder=1)    #绘制凸面图
axes.plot(X,Y,Z,'ro:',zorder=3)    #绘制z点变化的轨迹
plt.show()
```
