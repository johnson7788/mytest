```python
import random
import numpy as np
import matplotlib.pyplot as plt
#随机产生1-6，代表骰子的6个面
gen_ran = lambda : random.randint(1,6)

# 投掷n次产生的骰子点数之和
gen_men = lambda n: np.sum([gen_ran() for i in range(n)])
```

```python
#尝试多少次事件
num = 100000

#每次掷几次骰子
count = 3

#记录所有结果
res = [gen_men(3) for i in range(num)]
```

```python
#统计所有结果出现的次数
result = {k:res.count(k) for k in set(res)}
```

```python
x = sorted(result.keys())
y = [result[i]/num for i in x]
y
```

```python
plt.plot(x,y)
plt.grid(True)
plt.xlim(x[0]-1, x[-1]+1)
plt.xlabel('Sum of every event')
plt.ylabel('possibility')
plt.show()
```
