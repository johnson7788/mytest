```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#总共生成多少个随机数
number = 8000
#总共8000个1-9的随机数
A = np.random.randint(1,9,size=number)
#所有的x
x = list(range(number))
#从1取到8000个数，取前n个数的均值，得到y
y = [np.mean(A[0:i+1]) for i in range(number)]
```

```python
plt.plot(x,y)
plt.xlim(0,number)
plt.grid(True)
plt.show()
```
